```python
# step 1 is to load up the right libraries
# although this can vary we will probably use pandas, numpy and matplotlib 
#for matplotib, let's not take the whole thing, just pyplot

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
#import PIL
#from PIL import Image
```

Next we will go to the website https://covid.ourworldindata.org/ , get the data and print a few lines of our dataframe. 

```python
# let's see how many cases there are total recently

df = pd.read_csv('https://covid.ourworldindata.org/data/ecdc/total_cases.csv')
dftoday= df.fillna(0)
dftoday['day']= pd.DatetimeIndex(dftoday['date']).day
dftoday['month']= pd.DatetimeIndex(dftoday['date']).month
dftoday['year']= pd.DatetimeIndex(dftoday['date']).year
#print(dftoday.tail(5))
```

```python
# let's see how many cases there are total recently
 
#print(dftoday.tail(5))

dftoday.tail(5).style.set_properties(**{'background-color': 'white',                                                   
                                    'color': 'green',                       
                                    'border-color': 'white'})

```

```python
#Technical details on the whole csv table
print('The column names on the table are countries or areas and they are:',dftoday.columns)
print("The table has x entries with y data points- x,y here are:", dftoday.shape)
print("Let's see the types of data:", dftoday.dtypes)
```

```python
#let's double check that it's clean, in terms of duplicates
if dftoday.duplicated().any():
    print("This table has a duplicated row, so let me show it", dftoday.duplicated())
else:
    print("There are no duplicated rows")
    
```

```python
#check what the transposed dataframe looks like
df_transposed = dftoday.transpose()
print(df_transposed.head(3))
```

```python
# get the most recent date in the df
most_recent_date = dftoday['date'].max()
# get all the data on this most recent date
df_latest_day = dftoday[dftoday.date == most_recent_date]
most_recent_date_index = df_latest_day.columns[0]
#transpose
df_transposed_for_map = df_latest_day.transpose()
#print(df_transposed_for_map.head(5))
#get rid of date row
workable = df_transposed_for_map[2:-3]
print(workable.head(15))
print(workable.tail(15))
#workable.info()
len_minus_one = len(dftoday) - 1
print("The last row in the old df should have an index which now names the column and is",len_minus_one)
```

```python
#Check that the number is at top of dataframe
#if the two numbers below are the same- you are good to roll
print(df_transposed_for_map.columns[-1])
print(len_minus_one)
```

```python
#make this two columns and
#rename the columns properly

workable = workable.reset_index() 

workable = workable.rename(columns ={'index':'country', len_minus_one: 'count' })
print(workable.head(5))#workable 
workable.info()
```

```python
#transform the counts into an int
workable.count = workable['count'].astype(int)
workable.head(5)
```

```python
#Replacing the country name


workable_names = workable
workable_names.replace('United States', "United States of America", inplace = True)
workable_names.replace('Tanzania', "United Republic of Tanzania", inplace = True)
workable_names.replace('Democratic Republic of Congo', "Democratic Republic of the Congo", inplace = True)
workable_names.replace('Congo', "Republic of the Congo", inplace = True)
workable_names.replace('Serbia', "Republic of Serbia", inplace = True)
workable_names.replace('Czechia', "Czech Republic", inplace = True)
```

```python
folium_map = folium.Map(location=[40.738, -73.98], zoom_start=2)

#Setting up the world countries data URL
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'

country_shapes = f'{url}/world-countries.json'

folium.Choropleth(
    #The GeoJSON data to represent the world country
    geo_data=country_shapes,
    name='choropleth COVID-19',
    data=workable_names,
    #The column aceppting list with 2 value; The country name and  the numerical value
    columns=['country', 'count'],
    key_on='feature.properties.name',
    fill_color='PuBuGn',
    nan_fill_color='black'
).add_to(folium_map)
#for place in workable.country:
     
  #     workable.query('country==place')['count'] 
    #
 #   confirmed = workable_names.loc[workable_names['country']==place,'count']
   # df.loc[df['B'] == 3, 'A']
    # CONFIRMED cases
    # draw a circle on thecountry
  #  marker_confirmed = folium.CircleMarker(location=['country'],
   #                             radius=confirmed,
    #                            color="green",
     #                            fill=True)
   # marker_confirmed.add_to(folium_map)
    
folium_map.save("Covid-19 Map.html")
folium_map
```

```python
# covid-testing-all-observations.csv 
df_testing = pd.read_csv('https://covid.ourworldindata.org/data/testing/covid-testing-all-observations.csv')
df_testing_nonan= df_testing.fillna(0)
print(df_testing_nonan.head(1))
print(df_testing_nonan.tail(1))
df_testing_nonan.info()
```

```python

simplified_testing_dataframe= df_testing_nonan[['Entity','Date','Cumulative total']]  
simplified_testing_dataframe =simplified_testing_dataframe.rename(columns ={'Entity':'country', 'Cumulative total':'count' })

def make_country_name(name):
    if name.split()[0] == 'United':
           return name.split()[1] 
    else:
        return name.split()[0]
    
def split_on_minus(name):
    return name.split(" - ")[0]
# needs exception for two named countries    
simplified_testing_dataframe.country = simplified_testing_dataframe.country.apply(split_on_minus)
#print(simplified_testing_dataframe.head(5))
print(simplified_testing_dataframe.tail(5))
```

```python
# get the most recent date in the df
most_recent_date_testing = simplified_testing_dataframe['Date'].max()
# get all the data on this most recent date
simplified_testing_dataframe_last_day = simplified_testing_dataframe[simplified_testing_dataframe.Date == most_recent_date_testing]
simplified_testing_dataframe_index= simplified_testing_dataframe.columns[0]

workable_testing = simplified_testing_dataframe_last_day
#print(workable_testing.head(15))

#workable.info()
two_column_testing = simplified_testing_dataframe_last_day[['country','count']]
print("Only some countries report testing every day, today", len(two_column_testing), "reported testing")
legnth = len(two_column_testing)
print("We can see the list below\n")
print(two_column_testing.tail(legnth))
```

```python
top_testing_countries = two_column_testing.sort_values(by=['count'], ascending= False).head(20)
top_testing_countries['count_divided'] =top_testing_countries['count']/1000
#Visualization
#Create a new figure and make the size (12,10)
plt.figure(figsize=(12,10))
sns.barplot(x=top_testing_countries['country'], y=top_testing_countries['count_divided'],palette=sns.cubehelix_palette(20, start=2, rot=0, dark=0, light=.95, reverse=True))
# Place the country names at a 90-degree angle.
plt.xticks(rotation= 90)
plt.xlabel('Country')
plt.ylabel('Test Count in the thousands')
plt.title('Test Counts in 20 countries with highest cumulative test counts (divded by 1000) reporting today')
plt.show()
```
